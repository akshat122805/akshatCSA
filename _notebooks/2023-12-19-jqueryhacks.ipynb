<script>
  const oceanData = [
    { id: 1, name: 'Coral', email: 'coral@example.com' },
    { id: 2, name: 'Marlin', email: 'marlin@example.com' }
  ];

  function renderOceanData(data) {
    const tableBody = $('#ocean-table tbody');
    tableBody.empty();

    data.forEach(item => {
      const row = `
        <tr>
          <td>${item.id}</td>
          <td>${item.name}</td>
          <td>${item.email}</td>
          <td>
            <button class="edit-btn" data-id="${item.id}">Edit</button>
            <button class="remove-btn" data-id="${item.id}">Remove</button>
          </td>
        </tr>
      `;
      tableBody.append(row);
    });
  }

  function addOceanCharacter() {
    const newName = prompt('Enter the name of the Ocean character:');
    const newEmail = prompt('Enter the email of the Ocean character:');
    const newId = oceanData.length + 1;
    
    const newData = [...oceanData, { id: newId, name: newName, email: newEmail }];
    renderOceanData(newData);
  }

  $('#add-character-btn').on('click', addOceanCharacter);

  $('#ocean-table').on('click', '.remove-btn', function() {
    const idToRemove = $(this).data('id');
    const newData = oceanData.filter(item => item.id !== idToRemove);
    renderOceanData(newData);
  });

  $('#ocean-table').on('click', '.edit-btn', function() {
    const idToEdit = $(this).data('id');
    const editIndex = oceanData.findIndex(item => item.id === idToEdit);
    const newName = prompt('Enter a new name:');
    const newEmail = prompt('Enter a new email:');
    oceanData[editIndex] = { id: idToEdit, name: newName, email: newEmail };
    renderOceanData(oceanData);
  });

  // Initial rendering
  renderOceanData(oceanData);
</script>
